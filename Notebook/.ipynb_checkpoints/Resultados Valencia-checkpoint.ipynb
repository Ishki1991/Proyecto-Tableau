```python
import pandas as pd
```

```python
df_Valencia = pd.read_excel("datos/Partidos_del_Valencia.xlsx")
df_Valencia.head()
```

```python
df_Valencia['Fecha'] = df_Valencia['Fecha'].str.replace('Abr', 'Apr')
df_Valencia['Fecha'] = df_Valencia['Fecha'].str.replace('Ago', 'Aug')
df_Valencia['Fecha'] = df_Valencia['Fecha'].str.replace('Dic', 'Dec')
df_Valencia['Fecha'] = df_Valencia['Fecha'].str.replace('Ene', 'Jan')
df_Valencia.head()
```

```python
def add_result_column(df):
    result = []
    for index, row in df.iterrows():
        score = row["Score"]
        equipo1 = row["Equipo1"]
        equipo2 = row["Equipo2"]
        if "Valencia" in [equipo1, equipo2]:
            score1, score2 = map(int, score.split("-"))
            if equipo1 == "Valencia":
                if score1 > score2:
                    result.append("Ganó")
                elif score1 < score2:
                    result.append("Perdió")
                else:
                    result.append("Empató")
            else:
                if score2 > score1:
                    result.append("Ganó")
                elif score2 < score1:
                    result.append("Perdió")
                else:
                    result.append("Empató")
    df["Resultado"] = result
    return df

df_Valencia = add_result_column(df_Valencia)
```

```python
df_Valencia["Fecha"] = pd.to_datetime(df_Valencia["Fecha"])
df_Valencia
```

```python
df_Valencia["Score"].str.split(expand=True)

name = df_Valencia["Score"].str.split(expand=True)
name.columns = ['Goles_equipo1', "guion", 'Goles_equipo2']

df_Valen = pd.concat([df_Valencia, name], axis=1)
df_Valen
```

```python
df_Valen = df_Valen.drop(columns = ["guion"])
df_Valen["Goles_equipo1"] = df_Valen["Goles_equipo1"].astype(int)
df_Valen["Goles_equipo2"] = df_Valen["Goles_equipo2"].astype(int)

df_Valen
```

```python
df_Valen["Diferencia_goles"] = abs(df_Valen["Goles_equipo1"] - df_Valen["Goles_equipo2"])
df_Valen
```

```python
df_Valen.to_csv("data/df_partidos_Valencia.csv", index=False)
```
