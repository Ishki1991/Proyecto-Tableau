```python
import pandas as pd
```

```python
df_atl_bilbao = pd.read_excel("datos/Partidos_del_Ateltic_de_bilbao.xlsx")
df_atl_bilbao.head(10)
```

```python
df_atl_bilbao['Fecha'] = df_atl_bilbao['Fecha'].str.replace('Abr', 'Apr')
df_atl_bilbao['Fecha'] = df_atl_bilbao['Fecha'].str.replace('Ago', 'Aug')
df_atl_bilbao['Fecha'] = df_atl_bilbao['Fecha'].str.replace('Dic', 'Dec')
df_atl_bilbao['Fecha'] = df_atl_bilbao['Fecha'].str.replace('Ene', 'Jan')
df_atl_bilbao.head(10)
```

```python
df_atl_bilbao["Fecha"] = pd.to_datetime(df_atl_bilbao["Fecha"])
```

```python
df_atl_bilbao.head()
```

```python
df_atl_bilbao["Score"][0][0]
```

```python
def add_result_column(df):
    result = []
    for index, row in df.iterrows():
        score = row["Score"]
        equipo1 = row["Equipo1"]
        equipo2 = row["Equipo2"]
        if "Athletic" in [equipo1, equipo2]:
            score1, score2 = map(int, score.split("-"))
            if equipo1 == "Athletic":
                if score1 > score2:
                    result.append("Ganó")
                elif score1 < score2:
                    result.append("Perdió")
                else:
                    result.append("Empató")
            else:
                if score2 > score1:
                    result.append("Ganó")
                elif score2 < score1:
                    result.append("Perdió")
                else:
                    result.append("Empató")
    df["Resultado"] = result
    return df

df_atl_bilbao = add_result_column(df_atl_bilbao)
```

```python
df_atl_bilbao.head()
```

```python
df_atl_bilbao["Score"].str.split(expand=True)
```

```python
name = df_atl_bilbao["Score"].str.split(expand=True)
name.columns = ['Goles_equipo1', "guion", 'Goles_equipo2']

df_bilbao = pd.concat([df_atl_bilbao, name], axis=1)
df_bilbao
```

```python
df_bilbao = df_bilbao.drop(columns = ["guion"])
df_bilbao["Goles_equipo1"] = df_bilbao["Goles_equipo1"].astype(int)
df_bilbao["Goles_equipo2"] = df_bilbao["Goles_equipo2"].astype(int)
df_bilbao
```

```python
df_bilbao["Diferencia_goles"] = abs(df_bilbao["Goles_equipo1"] - df_bilbao["Goles_equipo2"])
df_bilbao
```

```python
df_bilbao.to_csv("data/df_partidos_bilbao.csv", index=False)
```
