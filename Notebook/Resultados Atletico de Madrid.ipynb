```python
import pandas as pd
```

```python
df_atl_madrid = pd.read_excel("datos/Partidos_del_Ateltico_de_Madrid.xlsx")
df_atl_madrid.head()
```

```python
df_atl_madrid['Fecha'] = df_atl_madrid['Fecha'].str.replace('Abr', 'Apr')
df_atl_madrid['Fecha'] = df_atl_madrid['Fecha'].str.replace('Ago', 'Aug')
df_atl_madrid['Fecha'] = df_atl_madrid['Fecha'].str.replace('Dic', 'Dec')
df_atl_madrid['Fecha'] = df_atl_madrid['Fecha'].str.replace('Ene', 'Jan')
df_atl_madrid.head()
```

```python
df_atl_madrid["Fecha"] = pd.to_datetime(df_atl_madrid["Fecha"])
```

```python
df_atl_madrid
```

```python
def add_result_column(df):
    result = []
    for index, row in df.iterrows():
        score = row["Score"]
        equipo1 = row["Equipo1"]
        equipo2 = row["Equipo2"]
        if "Atlético" in [equipo1, equipo2]:
            score1, score2 = map(int, score.split("-"))
            if equipo1 == "Atlético":
                if score1 > score2:
                    result.append("Ganó")
                elif score1 < score2:
                    result.append("Perdió")
                else:
                    result.append("Empató")
            else:
                if score2 > score1:
                    result.append("Ganó")
                elif score2 < score1:
                    result.append("Perdió")
                else:
                    result.append("Empató")
    df["Resultado"] = result
    return df

df_atl_madrid = add_result_column(df_atl_madrid)
```

```python
df_atl_madrid.head()
```

```python
df_atl_madrid["Score"].str.split(expand=True)
```

```python
name = df_atl_madrid["Score"].str.split(expand=True)
name.columns = ['Goles_equipo1', "guion", 'Goles_equipo2']

df_atlR = pd.concat([df_atl_madrid, name], axis=1)
df_atlR
```

```python
df_atlR = df_atlR.drop(columns = ["guion"])
df_atlR["Goles_equipo1"] = df_atlR["Goles_equipo1"].astype(int)
df_atlR["Goles_equipo2"] = df_atlR["Goles_equipo2"].astype(int)
```

```python
df_atlR
```

```python
df_atlR["Diferencia_goles"] = abs(df_atlR["Goles_equipo1"] - df_atlR["Goles_equipo2"])
df_atlR
```

```python
df_atlR.to_csv("data/df_partidos_AtlMadrid.csv", index=False)
```
