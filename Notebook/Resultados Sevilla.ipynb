```python
import pandas as pd
```

```python
df_Sevilla = pd.read_excel("datos/Partidos_del_Sevilla.xlsx")
df_Sevilla.head()
```

```python
df_Sevilla['Fecha'] = df_Sevilla['Fecha'].str.replace('Abr', 'Apr')
df_Sevilla['Fecha'] = df_Sevilla['Fecha'].str.replace('Ago', 'Aug')
df_Sevilla['Fecha'] = df_Sevilla['Fecha'].str.replace('Dic', 'Dec')
df_Sevilla['Fecha'] = df_Sevilla['Fecha'].str.replace('Ene', 'Jan')
df_Sevilla.head()
```

```python
df_Sevilla["Fecha"] = pd.to_datetime(df_Sevilla["Fecha"])
```

```python
df_Sevilla
```

```python
def add_result_column(df):
    result = []
    for index, row in df.iterrows():
        score = row["Score"]
        equipo1 = row["Equipo1"]
        equipo2 = row["Equipo2"]
        if "Sevilla" in [equipo1, equipo2]:
            score1, score2 = map(int, score.split("-"))
            if equipo1 == "Sevilla":
                if score1 > score2:
                    result.append("Ganó")
                elif score1 < score2:
                    result.append("Perdió")
                else:
                    result.append("Empató")
            else:
                if score2 > score1:
                    result.append("Ganó")
                elif score2 < score1:
                    result.append("Perdió")
                else:
                    result.append("Empató")
    df["Resultado"] = result
    return df

df_Sevilla = add_result_column(df_Sevilla)
```

```python
df_Sevilla["Score"].str.split(expand=True)
```

```python
name = df_Sevilla["Score"].str.split(expand=True)
name.columns = ['Goles_equipo1', "guion", 'Goles_equipo2']

df_sevi = pd.concat([df_Sevilla, name], axis=1)
df_sevi
```

```python
df_sevi = df_sevi.drop(columns = ["guion"])
df_sevi["Goles_equipo1"] = df_sevi["Goles_equipo1"].astype(int)
df_sevi["Goles_equipo2"] = df_sevi["Goles_equipo2"].astype(int)
```

```python
df_sevi
```

```python
df_sevi["Diferencia_goles"] = abs(df_sevi["Goles_equipo1"] - df_sevi["Goles_equipo2"])
```

```python
df_sevi
```

```python
df_sevi.to_csv("data/df_partidos_Sevilla.csv", index=False)
```
