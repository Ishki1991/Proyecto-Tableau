```python
import pandas as pd
```

```python
df_barca = pd.read_excel("datos/Partidos_del_barcelona.xlsx")
df_barca.head(10)
```

```python
df_barca['Fecha'] = df_barca['Fecha'].str.replace('Abr', 'Apr')
df_barca['Fecha'] = df_barca['Fecha'].str.replace('Ago', 'Aug')
df_barca['Fecha'] = df_barca['Fecha'].str.replace('Dic', 'Dec')
df_barca['Fecha'] = df_barca['Fecha'].str.replace('Ene', 'Jan')
df_barca.head()
```

```python
df_barca["Fecha"] = pd.to_datetime(df_barca["Fecha"])
```

```python
df_barca
```

```python
def add_result_column(df):
    result = []
    for index, row in df.iterrows():
        score = row["Score"]
        equipo1 = row["Equipo1"]
        equipo2 = row["Equipo2"]
        if "Barcelona" in [equipo1, equipo2]:
            score1, score2 = map(int, score.split("-"))
            if equipo1 == "Barcelona":
                if score1 > score2:
                    result.append("Ganó")
                elif score1 < score2:
                    result.append("Perdió")
                else:
                    result.append("Empató")
            else:
                if score2 > score1:
                    result.append("Ganó")
                elif score2 < score1:
                    result.append("Perdió")
                else:
                    result.append("Empató")
    df["Resultado"] = result
    return df

df_barca = add_result_column(df_barca)
```

```python
df_barca
```

```python
df_barca["Score"].str.split(expand=True)
```

```python
name = df_barca["Score"].str.split(expand=True)
name.columns = ['Goles_equipo1', "guion", 'Goles_equipo2']

df_barcelona = pd.concat([df_barca, name], axis=1)
df_barcelona
```

```python
df_barcelona = df_barcelona.drop(columns = ["guion"])
df_barcelona["Goles_equipo1"] = df_barcelona["Goles_equipo1"].astype(int)
df_barcelona["Goles_equipo2"] = df_barcelona["Goles_equipo2"].astype(int)
```

```python
df_barcelona["Diferencia_goles"] = abs(df_barcelona["Goles_equipo1"] - df_barcelona["Goles_equipo2"])
df_barcelona
```

```python
df_barcelona.to_csv("data/df_partidos_barcelona.csv", index=False)
```
