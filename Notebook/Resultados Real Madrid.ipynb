```python
import pandas as pd
```

```python
df_Real_Madrid = pd.read_excel("datos/Partidos_del_Real_Madrid.xlsx")
df_Real_Madrid.head()
```

```python
df_Real_Madrid['Fecha'] = df_Real_Madrid['Fecha'].str.replace('Abr', 'Apr')
df_Real_Madrid['Fecha'] = df_Real_Madrid['Fecha'].str.replace('Ago', 'Aug')
df_Real_Madrid['Fecha'] = df_Real_Madrid['Fecha'].str.replace('Dic', 'Dec')
df_Real_Madrid['Fecha'] = df_Real_Madrid['Fecha'].str.replace('Ene', 'Jan')
df_Real_Madrid.head()
```

```python
df_Real_Madrid["Fecha"] = pd.to_datetime(df_Real_Madrid["Fecha"])
```

```python
def add_result_column(df):
    result = []
    for index, row in df.iterrows():
        score = row["Score"]
        equipo1 = row["Equipo1"]
        equipo2 = row["Equipo2"]
        if "Real Madrid" in [equipo1, equipo2]:
            score1, score2 = map(int, score.split("-"))
            if equipo1 == "Real Madrid":
                if score1 > score2:
                    result.append("Ganó")
                elif score1 < score2:
                    result.append("Perdió")
                else:
                    result.append("Empató")
            else:
                if score2 > score1:
                    result.append("Ganó")
                elif score2 < score1:
                    result.append("Perdió")
                else:
                    result.append("Empató")
    df["Resultado"] = result
    return df

df_Real_Madrid = add_result_column(df_Real_Madrid)
```

```python
name = df_Real_Madrid["Score"].str.split(expand=True)
name.columns = ['Goles_equipo1', "guion", 'Goles_equipo2']

df_Rmadrid = pd.concat([df_Real_Madrid, name], axis=1)
df_Rmadrid
```

```python
df_Rmadrid = df_Rmadrid.drop(columns = ["guion"])
df_Rmadrid["Goles_equipo1"] = df_Rmadrid["Goles_equipo1"].astype(int)
df_Rmadrid["Goles_equipo2"] = df_Rmadrid["Goles_equipo2"].astype(int)
```

```python
df_Rmadrid
```

```python
df_Rmadrid["Diferencia_goles"] = abs(df_Rmadrid["Goles_equipo1"] - df_Rmadrid["Goles_equipo2"])
df_Rmadrid
```

```python
df_Rmadrid.to_csv("data/df_partidos_RealMadrid.csv", index=False)
```
